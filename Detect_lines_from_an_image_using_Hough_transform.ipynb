OpenCV function, ***cv2.HoughLines()***. It simply returns an array of (ρ,ϴ) values where ρ is measured in pixels and ϴ is measured in radians.

**Input Image**

parking_2.jpeg

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
img = cv2.imread("/content/gdrive/MyDrive/parking_2.jpeg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray, 75, 150)
lines = cv2.HoughLinesP(edges, 1, np.pi/180, 30, maxLineGap=250)
for line in lines:
   x1, y1, x2, y2 = line[0]
   cv2.line(img, (x1, y1), (x2, y2), (0, 0, 128), 1)
cv2_imshow(edges) #linesEdges
```
