```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
img=cv2.imread('/content/drive/MyDrive/coin',0)
plt.imshow(img)
```

```python
edges = cv2.Canny(img,100,200)
plt.imshow(edges)
```

```python
img.shape
```

```python
p1,p2=[],[]
for i in range(len(img)-1,-1,-1):
  #print(i,end=' ')
  for j in range(len(img[i])-1,-1,-1):
    #print(j,end=' ')
    if img[j][i]<200:
      p1=(i,j)
      break
  if not p1==[]:
    break

for i in range(len(img)):
  for j in range(len(img[i])-1,-1,-1):
    if img[j][i]<245:
      p2=(i,j)
      break
  if not p2==[]:
    break

p1,p2
```

```python
p2=(p2[0],p1[1])
cv2.line(img,p1,p2,(0,0,255),2)
plt.imshow(img)
```

```python
inches = 60     # 1 inch = 60 pixels 
pixel_count = abs(p1[0]-p2[0])
length = pixel_count/inches
length
```
